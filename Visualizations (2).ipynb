```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn
from matplotlib.pyplot import figure
from matplotlib import rcParams
from pylab import *
```

```python
iris = sbn.load_dataset("iris")
```

```python
sbn.boxplot(x="species", y="sepal_length",
             palette= "Blues",
            data= iris)
sbn.despine(offset=10, trim=True)
plt.title("sepal length density by species")
plt.grid()
plt.show()
```

```python
sbn.lineplot(x="sepal_length", y="sepal_width",
             hue="species",
             data=iris)
plt.title('Sepal length and width by species')
plt.show()
```

```python

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sbn.despine(f, left=True, bottom=True)
hue_order = ["yes","no"]
sbn.scatterplot(x="age", y="fare",
                  hue= "alive",
                palette="ch:r=-.4,d=.4_r",
                hue_order= hue_order,
                #sizes=(1, 8), linewidth=0,
                data=titanic, ax=ax)
plt.title("Distance and fare by number of passengers")
```
